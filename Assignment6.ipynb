```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mtplot
import itertools
import seaborn as sns

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB,GaussianNB,MultinomialNB
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from sklearn.decomposition import SparsePCA
```

```python
text_arr=[]
languages_arr=[]

with open("train.txt", mode='r', encoding="utf-8") as f:
    line = f.readline()
    while line:
        content,lang = line.split("\t")
        text_arr.append(content)
        languages_arr.append(lang[:-1])
        line = f.readline()
```

```python
dataframe = pd.DataFrame()
dataframe["text"]=text_arr
dataframe["lang"]=languages_arr
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(dataframe["text"])
y = dataframe["lang"]
```

```python
tsvd = TruncatedSVD(n_components=40)
kbest = SelectKBest(chi2, k=40)
combined_features = FeatureUnion([("svd", tsvd), ("chi2", kbest)])
X_features = combined_features.fit_transform(X,y)
```

```python
tsvd20 = TruncatedSVD(n_components=20)
tsvd40 = TruncatedSVD(n_components=40)
tsvd80 = TruncatedSVD(n_components=80)
```

```python
x_20 = tsvd20.fit_transform(X,y)
x_40 = tsvd40.fit_transform(X,y)
x_80 = tsvd80.fit_transform(X,y)
```

```python
sns.set(style="ticks", color_codes=True)
x=tsvd20.fit_transform(X,y)
dataframe = pd.DataFrame(x)
dataframe["target"]=y
df_shuffle = dataframe.sample(frac=1)
dataframe_new = df_shuffle.iloc[:100,:-1]
g = sns.pairplot(dataframe_new, palette="husl",kind="reg")
mtplot.savefig("graph1.png")
mtplot.show()
```

```python
X_temp = x_20
```

```python
def plot_graph(cm, classes, normalize=False, title='Confusion matrix', cmap=mtplot.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    else:
        print('Confusion matrix, without normalization')
    mtplot.figure(figsize=(15,10))
    mtplot.imshow(cm, interpolation='nearest', cmap=cmap)
    mtplot.title(title)
    mtplot.colorbar()
    tick_marks = np.arange(len(classes))
    mtplot.xticks(tick_marks, classes, rotation=45)
    mtplot.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        mtplot.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    mtplot.tight_layout()
    mtplot.ylabel('True label')
    mtplot.xlabel('Predicted label')
```

```python
X_temp = x_40
rf = RandomForestClassifier()
rf.fit(X_temp,y)
print(accuracy_score(y,rf.predict(X_temp)))
cnf_matrix = confusion_matrix(y,rf.predict(X_temp))
plot_graph(cnf_matrix,classes=np.unique(y),normalize=True,title = "Random Forest Confusion Matrix")
mtplot.savefig("graph3.png")
mtplot.show()
```

```python
lr = LogisticRegression()
lr.fit(X_temp,y)
print(accuracy_score(y,lr.predict(X_temp)))
cnf_matrix = confusion_matrix(y,lr.predict(X_temp))
plot_graph(cnf_matrix, classes=np.unique(y), normalize=True, title = "Logistic Regression Confusion Matrix")
mtplot.savefig("graph2.png")
mtplot.show()
```

```python
kfeatures = SelectKBest(chi2,k=80)
kbest_mn = kfeatures.fit_transform(X,y)
X_temp = kbest_mn
multinomail = MultinomialNB()
multinomail.fit(X_temp,y)
print(accuracy_score(y,multinomail.predict(X_temp)))
cnf_matrix = confusion_matrix(y,multinomail.predict(X_temp))
plot_graph(cnf_matrix,classes=np.unique(y),normalize=True, title = "Multinomial Confusion Matrix")
mtplot.savefig("graph5.png")
mtplot.show()
```

```python
X_temp=x_80
bernouli = BernoulliNB(alpha=1.0, binarize=0.0,class_prior=None, fit_prior=True)
bernouli.fit(X_temp,y)
print(accuracy_score(y,bernouli.predict(X_temp)))
cnf_matrix = confusion_matrix(y,bernouli.predict(X_features))
plot_graph(cnf_matrix,classes=np.unique(y),normalize=True,title = "Bernoulli Confusion Matrix")
mtplot.savefig("graph4.png")
mtplot.show()
```
